```python
import pandas as pd
```

```python
df = pd.read_csv('../data/recidivism.csv')
```

```python
df.columns
```

```python
df = df[['Sex','Release Type','Race - Ethnicity','Offense Classification','Offense Type','Offense Subtype','Return to Prison']]
```

```python
df = df[df.Sex=='Male']
df.drop('Sex',inplace=True,axis=1)
```

```python
for col in df.columns:
    print(df[col].value_counts(),'\n\n')
```

## Clean Null Rows

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
newdf = pd.get_dummies(df[['Release Type','Offense Classification',
       'Offense Type', 'Offense Subtype']])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encodeRecidivism = {'No': -1, 'Yes': 1}
```

```python
def encodeRace(x):
    if 'Black' in x:
        return 1
    else:
        return -1
```

```python
def encodeRecidivism(x):
    if x == 'No':
        return -1
    elif x == 'Yes':
        return 1
    else:
        return 0
```

```python
newdf['Race'] = df['Race - Ethnicity'].apply(lambda x: encodeRace(x))
```

```python
newdf['Recidivism'] = df['Return to Prison'].apply(lambda x: encodeRecidivism(x))
```
